**Part 2**

With the Product URL received in Part 1, extracting details:
• ASIN
• Product Description
• Manufacturer


```python
# Import the necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
# Read the data from a CSV file and store it in a DataFrame named 'products'
products = pd.read_csv("product_details.csv")
```

```python
products
```

```python
p_url = list(products['URL'])  # loading the urls of the products
```

```python
# Initialize empty lists to store product information
p_description = []  # To store product descriptions
p_ASIN = []         # To store ASIN (Amazon Standard Identification Number)
p_manufacturer = []  # To store the manufacturer information

# Loop through each product URL
for url in p_url:
    # Create a new WebDriver (Chrome browser in this case)
    driver = webdriver.Chrome()
    driver.get(url)

    time.sleep(10)  # Wait for the page content to load

    try:
        try:
            # Attempt to locate the product description element by ID "productDescription"
            product_description_element = driver.find_element(By.ID, "productDescription")
            product_description = product_description_element.text
        except:
            # If not found, try to locate the element by ID "aplus_feature_div"
            product_description_element = driver.find_element(By.ID, "aplus_feature_div")
            product_description = product_description_element.text
    except:
        # If no product description is found, set it to "N/A"
        product_description = "N/A"

    try:
        # Try to locate the product details element by ID "detailBullets_feature_div"
        product_details_element = driver.find_element(By.ID, "detailBullets_feature_div")
        product_detail = product_details_element.text
        # Split the product details by newlines and look for "Manufacturer" and "ASIN" information
        for data in product_detail.split("\n"):
            if "Manufacturer" in data:
                manufacturer = data.split(":")[1]
            elif "ASIN" in data:
                asin = data.split(":")[1]
    except:
        # If no product details are found, set ASIN and manufacturer to "N/A"
        asin = "N/A"
        manufacturer = "N/A"

    # Append the extracted information to respective lists
    p_description.append(product_description)
    p_ASIN.append(asin)
    p_manufacturer.append(manufacturer)

    # Quit the WebDriver for this product URL
    driver.quit()
```

```python
# Create a DataFrame 'df_more' using data from lists 'p_ASIN', 'p_manufacturer', and 'p_description'
df_more = pd.DataFrame({'ASIN':p_ASIN,'manufacturer':p_manufacturer,'description':p_description})
```

```python
df_more
```

```python
# Extract the 'ASIN' column from DataFrame 'df_more' into a list 'p_ASIN'
p_ASIN = list(df_more['ASIN'])

# Extract the 'manufacturer' column from DataFrame 'df_more' into a list 'p_manufacturer'
p_manufacturer = list(df_more['manufacturer'])

# Extract the 'description' column from DataFrame 'df_more' into a list 'p_description'
p_description = list(df_more['description'])
```

```python
# Add a new column 'ASIN' to the 'products' DataFrame and populate it with the values from the 'p_ASIN' list
products['ASIN'] = p_ASIN

# Add a new column 'Description' to the 'products' DataFrame and populate it with the values from the 'p_description' list
products['Description'] = p_description

# Add a new column 'Manufacturer' to the 'products' DataFrame and populate it with the values from the 'p_manufacturer' list
products['Manufacturer'] = p_manufacturer
```

```python
products
```

```python
# Drop rows with NaN values from the 'products' DataFrame
products = products.dropna()
```

```python
# Save the 'products' DataFrame to a CSV file named 'final_product_details.csv' without including the index column
products.to_csv('final_product_details.csv',index=False)
```
